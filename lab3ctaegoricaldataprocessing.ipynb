## Empirical Distribution for Categorical Data

```python
cat_data = ['A','B','A','C','B','A','D','B','C','A']
```

```python
# Frequency table
import pandas as pd

s= pd.Series(cat_data)

freq = s.value_counts()
print(freq)
```

```python
import matplotlib.pyplot as plt
```

```python
# Bar Chart
freq.plot(kind='bar')
plt.title("Bar Chart - Categorical Data")
plt.xlabel("Category")
plt.ylabel("Count")
plt.show()
```

```python
# Pareto Chart
# Bars show frequency: line shows cumulative percentage

cum = freq.cumsum()/ freq.sum()

fig, ax1 = plt.subplots()
ax1.bar(freq.index, freq)
ax2 = ax1.twinx()
ax2.plot(freq.index, cum, marker ='o')
plt.title("Pareto Chart")
plt.show()
```

```python
# Pie Chart
# Shows proportion of each category in a circular chart
s.value_counts().plot(kind='pie',autopct='%1.1f%%')
plt.title('Pie Chart')
plt.ylabel('')
plt.show()
```
