Lab 2:Plotting and inferring empirical distribution

```python
import numpy as np
new_data = np.array([30, 32, 29, 35, 40, 42, 38, 36, 34, 31, 33, 37, 39])
```

```python
# Histogram
# Shows the frequency of values grouped into bins; approximates the empirical PDF.
import matplotlib.pyplot as plt
plt.hist(new_data, bins=5, density=True)
plt.title("Histogram - Numerical Data")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
# Empirical CDF
# Description: Shows cumulative

num_data = np.array([5, 8, 7, 6, 9, 12, 15, 14, 11, 10, 13, 16, 18, 17])
```

```python
x = np.sort(num_data)
y = np.arange(1, len(num_data)+1) / len(num_data)

plt.plot(x,y)
plt.title("ECDF - Numerical Data")
plt.xlabel("Value")
plt.ylabel("Cummulative Probability")
plt.show()
```

```python
from scipy.stats import gaussian_kde

kde = gaussian_kde(num_data)
x_vals = np.linspace(min(num_data), max(num_data), 100)
y_vals = kde(x_vals)

plt.plot(x_vals, kde(x_vals))
plt.title("Kernel Density Estimation (KDE)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
# Box Plot
# Shows median, quartiles and outliers
plt.boxplot(num_data)
plt.title("Box Plot - Numerical Data")
plt.ylabel("Value")
plt.show()
```

```python
# Violin Plot

plt.violinplot(num_data)
plt.title("Violin Plot - Numerical Data")
plt.ylabel("Value")
plt.show()
```
